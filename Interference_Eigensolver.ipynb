# Interference Eigensolver

We will use the face that the ground state oscilates at the slowest frequency.  Because of this, if we add together many arbitrary time evolution opperators, the ground state will be the last on to average to zero.  So how do we add time evolution operators on a quantum computer?

We know how to implement
$$U_I(t) = (I-i dt H\otimes I)^\tau$$
$$U_Z(t) = (I-i dt H\otimes Z)^\tau$$
where $t=\tau dt$.  So we can implement
$$ U_I(t_a)U_Z(t_b) = \begin{pmatrix} e^{-i H t_1} & 0 \\ 0 & e^{-i H t_2} \end{pmatrix}$$
where $t_1 = t_a + t_b$ and $t_2 = t_a - t_b$.  

Applied to the state $|\psi>\otimes(|0> + |1>)$ where $|\psi>$ is arbitrary
$$U_I(t_a)U_Z(t_b)|\psi>\otimes(|0> + |1>) = e^{-i H t_1}|\psi;0>+e^{-i H t_2}|\psi;1>$$

Now we can interfere these two time evolutions by an application of the Hadamard gate $O_H$.
$$O_H U_I(t_a) U_Z(t_b)|\psi>\otimes(|0> + |1>)  = (e^{-i H t_1}+e^{-i H t_2})|\psi;0> + (e^{-i H t_1}-e^{-i H t_2})|\psi;1>$$

Let us define
$$U(t_1,t_2) = O_H U_I(t_a) U_Z(t_b)$$
Then continually applying U(t_1,t_2) with arbitrary $t_1$, $t_2$ should take us to the ground state.

## Integration checks

Let's start with a santiy check.  We should have that $\sum_i e^{i \phi_i} \rightarrow 0$ 

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def Rphi():
    return random.randrange(2*314)/100

def S(N):
    return sum([np.exp(1j*Rphi()) for i in range(0,N)])

avg = []
nl = []
for n in range(1,1000):
    avg.append(S(n)/n)
    nl.append(n)
    
plt.plot(nl,avg)
```

What happens to the on qubit?  Let's do a test where $H=1$.

```python
def Mdot(M):
    out = M[0]
    for i in range(1,len(M)):
        out = np.dot(M[i],out)
    return out
```

```python
O_H = 1/np.sqrt(2)*np.array([[1,1],[1,-1]])
def U():
    return np.array([[np.exp(1j*Rphi()),0],[0,np.exp(1j*Rphi())]])
psi = 1/np.sqrt(2)*np.array([1,1])


```

```python
avg_0 = []
avg_1 = []
norm = []
nl = []
psi_f = psi
for n in range(0,300):
    psi_f = Mdot([O_H,U(),psi_f])
    avg_0.append(np.abs(psi_f[0])**2)
    avg_1.append(np.abs(psi_f[1])**2)
    norm.append(np.dot(np.conjugate(psi_f),psi_f))
    nl.append(n)
    
plt.plot(nl,avg_0)
plt.plot(nl,avg_1)
plt.plot(nl,norm)
```

It's not suprising that niether state decays to zero since the varience is one.  

Let's take $\phi: 0 \rightarrow 2\pi$

```python
O_H = 1/np.sqrt(2)*np.array([[1,1],[1,-1]])
def U(t1,t2):
    return np.array([[np.exp(1j*t1),0],[0,np.exp(1j*t2)]])
psi = 1/np.sqrt(2)*np.array([1,1])
```

```python
dt1 = 0.01
dt2 = 0.011

avg_0 = []
avg_1 = []
norm = []
nl = []
psi_f = psi
for n in range(0,300):
    psi_f = Mdot([O_H,U(dt1*n,dt2*n),psi_f])
    avg_0.append(np.abs(psi_f[0])**2)
    avg_1.append(np.abs(psi_f[1])**2)
    norm.append(np.dot(np.conjugate(psi_f),psi_f))
    nl.append(n)
    
plt.plot(nl,avg_0)
plt.plot(nl,avg_1)
plt.plot(nl,norm)
```
