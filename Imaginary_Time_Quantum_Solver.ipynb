```python
from qiskit import quantum_info as qi
import numpy as np
import pandas as pd
import scipy.linalg as lng
pd.options.display.float_format = '{:,.1f}'.format
```

```python
def bkt(psi_a,psi_b):
    return np.dot(np.conjugate(psi_a),psi_b)

def exv(psi_a,O,psi_b):
    return np.dot(np.conjugate(psi_a),np.dot(O,psi_b))

def Mdot(M):
    mtp = M[0]
    for i in range(1,len(M)):
        mtp = np.dot(M[i],mtp)
    return mtp
```

```python
def I0(L):
    label = ""
    for i in range(0,L):
        label += "I"
    return qi.Operator.from_label(label).data

def Sx(l,L):
    label = ""
    for i in range(0,l):
        label += "I"
    label += "X"
    for i in range(l+1,L):
        label += "I"
    return qi.Operator.from_label(label).data

def Sy(l,L):
    label = ""
    for i in range(0,l):
        label += "I"
    label += "Y"
    for i in range(l+1,L):
        label += "I"
    return qi.Operator.from_label(label).data

def Sz(l,L):
    label = ""
    for i in range(0,l):
        label += "I"
    label += "Z"
    for i in range(l+1,L):
        label += "I"
    return qi.Operator.from_label(label).data
    
def Sp(l,L):
    return Sx(0,L)+Sy(0,L)

def Sm(l,L):
    return Sx(0,L)-Sy(0,L)
```

```python
def H(Jx,Jy,Jz,L):
    h=0
    for i in range(0,L-1):
        h+= Jx*np.dot(Sz(i,L),Sz(i+1,L))
        h+= Jy*np.dot(Sx(i,L),Sx(i+1,L))
        h+= Jz*np.dot(Sy(i,L),Sy(i+1,L))
    h+= Jx*np.dot(Sz(L-1,L),Sz(0,L))
    h+= Jy*np.dot(Sx(L-1,L),Sx(0,L))
    h+= Jz*np.dot(Sy(L-1,L),Sy(0,L))
    return h
        
```

```python
H1 = H(1,1,1,2)
(e,U) = lng.eig(H1)
psi = np.transpose(U)
exv(psi[1],H1,psi[1])
psi
```

```python
def Imaginary_Time_Solver(H,dt,n):
    L = int(np.log2(len(H)))
    Ui = I0(L) - H*dt
    psi_0 = np.array([0.1,0.2,0.3,0.4])
    psi_0 = psi_0/np.sqrt(np.dot(psi_0,psi_0))
    for i in range(0,n):
        psi_0 = np.dot(Ui,psi_0)
        psi_0 = psi_0/np.sqrt(np.dot(psi_0,psi_0))
    return psi_0
```

### The problem is that the imaginary time operator is not unitary

Can we find a similar operator that is unitary?  Let us consider that in nature the high energy states decay because they ineract faster with the enviroment.  To this extant we add a qubit to simulate the enviroment and make a new Hamiltonian $H_2 = H_1 X_n$ so that 
$$(1-iH_2 \sqrt{dt})(1-iH_2 \sqrt{dt}) = 1-H_1^2 dt - 2 i H_1 X_n \sqrt{dt} $$
The real time part is now moved to the occupied enviroment block and in the unoccupied block we have imaginary time evolution of the Hamiltonian $H_1^2$.  The question is, can we find an $a$ such that $(a+H_1)^2$ and $H_1$ have the same ground state?

Let $|e_n>$ be the eiganstates of $H_1$ with $e_0$ being the ground state energy.  Then
$$ (a+H_1)^2|e_n> = (a+e_n)^2 |e_n> $$
so as long as $a+e_0 > 0$ then $|e_0>$ is also the ground state of $(a+H_1)^2$.  

We can look at what is being applied in both sectors.  If we apply the discrete evolution a large number of times $\tau$ then we have
$$ (1-i dt H_1 X_n)^{\tau} = \sum_{k=0}^{\tau}\frac{\tau !}{k! (\tau-k)!}(-i dt H_1 X_n)^{k} \approx \sum_{k=0}^{\tau}\frac{1}{k!}(-i \tau dt H_1 X_n)^{k}$$
We can break this into even and odd values of k
$$ (1-i dt H_1 X_n)^{\tau} \approx \sum_{l=0}^{\tau/2}\frac{1}{(2l)!}(- (\tau dt)^2 H_1^2)^{l} + \sum_{l=0}^{\tau/2}\frac{1}{(2l+1)!}(- (\tau dt)^2 H_1^2 )^{l} (i H_1 X_n)$$
If we act with this on an arbitrary state which is empty in the extra qubit $|x;0>$ we see that we have a decaying function in both sectors.

$$ (1-i dt H_1 X_n)^{\tau}|x;0> \approx \sum_{l=0}^{\tau/2}(- (\tau dt)^2 H_1^2)^{l}\left(\frac{1}{(2l)!}|x;0> +  \frac{i}{(2l+1)!}H_1|x;1>\right) $$

This is not exactly $e^{- (H_1^ \tau dt)^2}$ because we have the fractions $\frac{1}{(2l)!}$ and $\frac{1}{(2l+1)!}$ instead of $\frac{1}{(l)!}$.  However, since these fractions are smaller the function should actually decay faster.  In fact we have

$$ (1-i dt H_1 X_n)^{\tau}|x;0> \approx \sum_{l=0}^{\tau/2}\left[\cos\left((\tau dt)^2 H_1^2\right)|x;0> +  i \sin\left((\tau dt)^2 H_1^2\right)|x;1>\right] $$

So it looks like we have to pick the right time in order to find the ground state.  This is like VQE with only one variational parameter $t$.  It also looks simillar to QAOA but it is certainly different.
